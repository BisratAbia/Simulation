# Mean axial velocity $U_z^+$

```python
sarmamean = pd.read_csv('C:/Users/pmbw/Development/datasets/particles/Vpipe180/partstat03500.txt',
           skiprows =1, sep='\s+', engine='python')
# partstatvreman1 = pd.read_csv('C:/Users/pmbw/Development/datasets/particles/Vpipe180/partstat03000.txt',
#            skiprows =1, sep='\s+', engine='python')
sarma = pd.read_csv('C:/Users/pmbw/Development/datasets/particles/restrted/sarmamean95Z.txt',
            skiprows=1, sep='\s+', engine='python') 
ax = plt.gca()
# Set x logaritmic
#ax.set_xscale('log')
#plt.xlim(0.01,1)
#plt.ylim(0,25)
#plt.plot((sarma.r)*2,sarma.ave/14.35229, color = 'black',linewidth=1, markersize=1)
plt.plot((1-sarmamean.y*2),sarmamean.w,color ='red', linewidth=1, markersize=1)
plt.plot((1-sarma.r*2),(sarma.ave/18.29329), color = 'black',linewidth=1, markersize=1)
plt.title('Mean particle profile - 360' , fontsize = 22)
plt.ylabel('$u_z^*$',fontsize=22)
plt.xlabel('$(1-r)^*$' ,fontsize=18)
plt.legend({'current run','Sarma DNS'},loc='lower right' )
```

```python
sarmarms = pd.read_csv('C:/Users/pmbw/Development/datasets/particles/restrted/sarmarms95Z.txt',
            skiprows=1, sep='\s+', engine='python') 
ax = plt.gca()
# Set x logaritmic
#ax.set_xscale('log')
#plt.xlim(0.01,1)
#plt.ylim(0,25)
#plt.plot((sarma.r)*2,sarma.ave/14.35229, color = 'black',linewidth=1, markersize=1)
plt.plot((1-sarmamean.y*2),np.sqrt(sarmamean.w2)*17.359,color ='red', linewidth=1, markersize=1)
plt.plot((1-sarmarms.r*2),(sarmarms.ave), color = 'black',linewidth=1, markersize=1)
plt.title('RMS - 360' , fontsize = 22)
plt.ylabel('$u_z^*$',fontsize=22)
plt.xlabel('$(1-r)^*$' ,fontsize=18)
plt.legend({'current run','Sarma DNS'})
```
